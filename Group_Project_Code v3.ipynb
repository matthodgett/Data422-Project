```r
# To start, we can slice of the first two rows as these are not needed, this information can be put in (to be created) read.me file
Pollution <- read_csv("Pollution.csv", col_names=FALSE)

Pollution <- Pollution %>%
    slice(3:nrow(Pollution))

Pollution %>%
    head()
```

```r
# Cool, but we still need to remove/make first row the column names
Pollution <- Pollution %>%
    rename(
        Country = X1,
        Cause = X2,
        Both_sexes = X3,
        Male = X4,
        Female = X5,
        Both_sexes_standardised = X6,
        Male_standardised = X7,
        Female_standardised = X8
    ) %>%
    slice(2:nrow(Pollution))

Pollution %>% head()
```

```r
# Idea: fix numbers first then any filtering and creating new data frames will use already fixed numbers?
# Will need to use regex and string_extract
Pollution <- Pollution %>%
    mutate(
        Both_sexes_value = str_extract(Both_sexes, "\\d+"),
        Both_sexes_lower = str_replace(str_extract(Both_sexes, "\\[ \\d+"), "[\\[]", ""), 
        Both_sexes_upper = str_replace(str_extract(Both_sexes, "\\–\\d+"), "–", ""),
        Male_value = str_extract(Male, "\\d+"),
        Male_lower = str_replace(str_extract(Male, "\\[ \\d+"), "[\\[]", ""), 
        Male_upper = str_replace(str_extract(Male, "\\–\\d+"), "–", ""),  
        Female_value = str_extract(Male, "\\d+"),
        Female_lower = str_replace(str_extract(Male, "\\[ \\d+"), "[\\[]", ""), 
        Female_upper = str_replace(str_extract(Male, "\\–\\d+"), "–", ""),
        Both_sexes_standardised_value = str_extract(Both_sexes_standardised, "\\d+"),
        Both_sexes_standardised_lower = str_replace(str_extract(Both_sexes_standardised, "\\[ \\d+"), "[\\[]", ""), 
        Both_sexes_standardised_upper = str_replace(str_extract(Both_sexes_standardised, "\\–\\d+"), "–", ""),
        Male_standardised_value = str_extract(Male_standardised, "\\d+"),
        Male_standardised_lower = str_replace(str_extract(Male_standardised, "\\[ \\d+"), "[\\[]", ""), 
        Male_standardised_upper = str_replace(str_extract(Male_standardised, "\\–\\d+"), "–", ""),
        Female_standardised_value = str_extract(Male_standardised, "\\d+"),
        Female_standardised_lower = str_replace(str_extract(Male_standardised, "\\[ \\d+"), "[\\[]", ""), 
        Female_standardised_upper = str_replace(str_extract(Male_standardised, "\\–\\d+"), "–", ""),
          )

Pollution %>% head()
#gsub("[\\[\\]]", "", regmatches(j, gregexpr("\\[.*?\\]", j))[[1]])
#str_replace(str_extract(Both_sexes, "\\[ \\d+ \\]"), "[\\[]", "")
#Both_sexes_upper = str_replace(str_extract(Both_sexes, "\\[.*?\\]"), "[\\[ \\d+.]", "")
```
